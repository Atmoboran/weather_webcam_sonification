```python
from functions.video_funcs_1 import generate_weather_animation, read_station_data, load_weather_and_image_data
from pathlib import Path

BASE_DIR = "C:/Users/Frank/Documents/python/weather_webcam_sonification/"
print('Base directory is:', BASE_DIR)
WEATHER_DATA_DIR = BASE_DIR + "weatherdata/"

# Define input weather data files
WEATHER_FILE_TEMP = WEATHER_DATA_DIR + "produkt_zehn_min_tu_20230828_20250227_07431.txt"
WEATHER_FILE_WIND = WEATHER_DATA_DIR + "produkt_zehn_min_ff_20230831_20250302_07341.txt"
WEATHER_FILE_PRECIP = WEATHER_DATA_DIR + "produkt_zehn_min_rr_20230831_20250302_07341.txt"

# Read the data
print("Loading weather data...")
df_temp = read_station_data(str(WEATHER_FILE_TEMP))
df_wind = read_station_data(str(WEATHER_FILE_WIND))
df_precip = read_station_data(str(WEATHER_FILE_PRECIP))

```

```python
# Ensure 'MESS_DATUM' exists and merge only common datetimes
if 'MESS_DATUM' in df_temp.columns and 'MESS_DATUM' in df_wind.columns and 'MESS_DATUM' in df_precip.columns:
    # Find common 'MESS_DATUM' values across all dataframes
    common_dates = set(df_temp['MESS_DATUM']) & set(df_wind['MESS_DATUM']) & set(df_precip['MESS_DATUM'])
    
    # Filter each dataframe to keep only the common dates
    df_temp_filtered = df_temp[df_temp['MESS_DATUM'].isin(common_dates)]
    df_wind_filtered = df_wind[df_wind['MESS_DATUM'].isin(common_dates)]
    df_precip_filtered = df_precip[df_precip['MESS_DATUM'].isin(common_dates)]
    
    # Merge dataframes on 'MESS_DATUM' (inner join ensures only common dates remain)
    df_merged = df_temp_filtered.merge(df_wind_filtered, on='MESS_DATUM', how='inner').merge(df_precip_filtered, on='MESS_DATUM', how='inner')
    
    # Determine the first and last timestamp for the filename and format them
    start_time = df_merged['MESS_DATUM'].min().strftime('%Y%m%d_%H%M')
    end_time = df_merged['MESS_DATUM'].max().strftime('%Y%m%d_%H%M')
    merged_file_path = WEATHER_DATA_DIR + f"merged_weather_data_{start_time}_{end_time}.txt"
    
    df_merged.to_csv(merged_file_path, sep=';', index=False)
    print(f"Merged weather data saved at: {merged_file_path}")
else:
    print("Error: 'MESS_DATUM' column not found in one or more dataframes.")
```
